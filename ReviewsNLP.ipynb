```python
def clean_text(df):
    custom_pipeline = [preprocessing.lowercase,
                  preprocessing.remove_punctuation,
                  preprocessing.remove_stopwords,
                  preprocessing.stem,
                  preprocessing.remove_whitespace]
    df['clean_text'] = hero.clean(df['reviewText'], custom_pipeline)
```

```python
clean_text(df_b_r)
```

```python
df_b_r
```

```python
df_b_r['tfidf'] = (hero.tfidf(df_b_r['clean_text'], max_features=50))
```

```python
df_b_r
```

```python
df_b_r['tsnetfidf'] = hero.tsne(df_b_r['tfidf'])
```

```python
hero.scatterplot(df_b_r, col='tsnetfidf', color='asin'
                 , title="TF-IDF", hover_data = ['asin','reviewText'])
```
